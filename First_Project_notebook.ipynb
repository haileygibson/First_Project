```python
df['Country'].value_counts()[["United States", "India", "Germany", "United Kingdom", "Canada"]]
```

```python
group_of_countries = df.groupby("Country").mean()
```

```python
group_of_countries["Salary_in_USD"]
```

```python
list_of_countries = ['Nepal', 'Mali', 'United States','United Kingdom','Canada','France']
```

### Top 5 Social Media Platforms in Nepal, Mali, United States, United Kingdom, Canada, and France

```python
for x in list_of_countries:
    print(x + "'s top 10 SocialMedia","\n",
          df[df['Country']== x]['SocialMedia'].value_counts().nlargest(10),
          "\n", "************************************")
```

### Top Salary(USD) in Nepal, Mali, United States, United Kingdom, Canada, and France¶

```python
for x in list_of_countries:
    print(x + "'s top 10 Salry_in_USD","\n",
          df[df['Country']== x]['Salary_in_USD'].value_counts().nlargest(10),
          "\n", "_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ ")
```

### Mean, Max, and Median Salary(USD) in Nepal, Mali, United States, United Kingdom, Canada, and France¶

```python
for x in list_of_countries:
    print(x,"\n",
          df[df['Country']== x]['Salary_in_USD'].agg(["mean", "max", "median"]),
          "\n", "_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ ", "\n")
```
